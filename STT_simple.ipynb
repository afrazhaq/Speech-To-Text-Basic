```python
import speech_recognition as sr
import pyaudio

init_rec = sr.Recognizer()
print("Let's speak!")

with sr.Microphone() as source:
 audio_data = init_rec.record(source, duration=10)
 print("Recognizing...")
 text = init_rec.recognize_google(audio_data)
 print(text)
```
